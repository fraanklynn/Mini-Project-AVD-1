## Visualisasi Data

### Histogram

Distribusi Lama Kerja Karyawan di Perusahaan

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["YearsAtCompany"], bins=20, kde=True, color="blue")
plt.title("Distribusi Lama Kerja Karyawan (YearsAtCompany)")
plt.xlabel("Tahun di Perusahaan")
plt.ylabel("Jumlah Karyawan")
plt.show()
```

INSIGHT
1. Mayoritas karyawan memiliki masa kerja kurang dari 5 tahun, terlihat dari puncak histogram.
2. Lonjakan jumlah karyawan di tahun ke-5 dan ke-10 bisa jadi akibat pola promosi atau kontrak kerja tertentu.
3. Penurunan yang signifikan jumlah karyawan setelah 10 tahun bisa ada indikasi kurangnya tunjangan, peluang kenaikan jabatan, atau penghargaan khusus bagi karyawan senior, yang dapat menyebabkan mereka mencari peluang di tempat lain.

TINDAKAN YANG DAPAT DIAMBIL
1. Evaluasi penyebab pengunduran diri karyawan di awal masa kerja dan berikan pelatihan.
2. Analisis kebijakan promosi untuk memastikan keberlanjutan tenaga kerja setelah 5 dan 10 tahun.
3. Menyediakan program pengembangan karier yang jelas, termasuk peluang promosi dan peningkatan keterampilan bagi karyawan senior agar mereka tetap termotivasi untuk bertahan di perusahaan.









### Bar Chart

 Menghitung jumlah karyawan yang mengundurkan diri dan yang tetap bekerja berdasarkan jabatan di perusahaan

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x="JobRole", hue="Attrition", palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Jumlah Karyawan yang Bertahan dan Mengundurkan Diri Berdasarkan Jabatan di Perusahaan")
plt.xlabel("Jabatan")
plt.ylabel("Jumlah Karyawan")
plt.legend(title="Pengunduran Diri", labels=["Tidak", "Ya"])
plt.show()
```

INSIGHT
1. Jabatan Sales Executive, Laboratory Technician, dan Manufacturing Director memiliki tingkat pengunduran diri yang tinggi. Hal ini menunjukkan bahwa karyawan pada posisi tersebut lebih rentan untuk mengundurkan diri dibandingkan dengan jabatan lainnya.
2. Jabatan seperti Research Director dan Manager memiliki tingkat pengunduran diri yang lebih rendah. Hal ini dapat mengindikasikan adanya stabilitas kerja yang lebih baik atau adanya faktor lain yang membuat karyawan bertahan lebih lama.

TINDAKAN YANG DAPAT DIAMBIL
1. Melakukan evaluasi terhadap faktor penyebab pengunduran diri pada jabatan
2. Meningkatkan kepuasan kerja dengan meninjau kembali beban kerja, peluang pengembangan karier, dan fleksibilitas kerja.
3. Memastikan rekrutmen dan seleksi yang lebih tepat agar kandidat memahami tuntutan pekerjaan sejak awal.

### Pie Chart

Membandingkan Karyawan yang Lembur dan tidak Lembur

```python
plt.figure(figsize=(6, 6))
colors = ['lightcoral', 'skyblue']
plt.pie(df['OverTime'].value_counts(),
        labels=df['OverTime'].value_counts().index,
        autopct='%1.1f%%',
        colors=colors)
plt.title('Proporsi Karyawan yang Lembur')
plt.show()
```

INSIGHT
1. Sebagian besar karyawan tidak melakukan lembur. Hal ini menunjukkan bahwa sebagian besar tugas atau beban kerja sudah seimbang dan dapat diselesaikan pada saat jam kerja
2. Sebagian kecil karyawan atau 28,2% karyawan melakukan lembur. Hal ini bisa jadi karena kebutuhan operasional tertentu, beban kerja tinggi, atau kurangnya sumber daya manusia

TINDAKAN YANG DAPAT DIAMBIL
1. Evaluasi kebijakan lembur jika lembur tidak merata di semua posisi, pertimbangkan untuk menyesuaikan beban kerja lebih seimbang.
2. Jika lembur diperlukan, pastikan ada kompensasi yang adil atau fleksibilitas kerja untuk menjaga keseimbangan kehidupan dan pekerjaan.

### Heatmap

Korelasi faktor faktor dengan pengunduran diri karyawan antar variabel numerik



```python
df_numeric = df.select_dtypes(include=['number'])
```

```python
plt.figure(figsize=(15, 20))
sns.heatmap(df_numeric.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap Korelasi Antar Variabel Numerik')
plt.show()
```

INSIGHT
1. Terdapat hubungan yang sangat kuat antara tingkat jabatan(Job Level) dan pendapatan bulanan(Monthly Income), yang menunjukkan bahwa kenaikan gaji terutama bergantung pada promosi jabatan.Hal ini menandakan bahwa peningkatan penghasilan hanya dapat dicapai melalui kenaikan jabatan.
2. Korelasi Kuat antara Lama Bekerja di Perusahaan (Years At Company) dan Lama dalam Jabatan Saat Ini (Years In Current Role). Banyak karyawan yang sudah lama bekerja di perusahaan tetapi tetap dalam jabatan yang sama. Hal ini bisa menunjukkan terbatasnya peluang promosi atau pengembangan karir.


TINDAKAN YANG DAPAT DIAMBIL
1. Melakukan evaluasi terhadap kebijakan kompensasi untuk memastikan bahwa sistem yang diterapkan sudah adil dan tidak menghambat karyawan yang memiliki kinerja tinggi tetapi belum mendapatkan promosi.
2. Melakukan evaluasi terhadap sistem promosi untuk memastikan bahwa kenaikan jabatan tidak hanya bergantung pada masa kerja, tetapi juga pada pencapaian kinerja.
